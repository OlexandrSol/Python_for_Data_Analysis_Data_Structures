```python
# Можна створити Series із словника
sdata = {'Ohio': 35000, 'Texas': 71000, 'Oregon': 16000, 'Utah': 5000}
```

```python
sdata
```

```python
obj3 = pd.Series(sdata)
```

```python
obj3
```

```python
obj3['Ohio']
```

```python
# Обернути назад в словник
obj3.to_dict()
```

```python
obj3
```

```python
states = ['California', 'Ohio', 'Oregon', 'Texas']
```

```python
obj4 = pd.Series(sdata, index=states)
```

```python
obj4
```

```python
pd.isna(obj4)
```

```python
pd.notna(obj4)
```

```python
# Метод екземпляра
obj4.isna()
```

```python
# Корисно, що дані вирівнюються по індексній мітці
obj3 + obj4
```

```python
# В самого індекса і об'єкта Series є атрибут name
obj4.name = 'population'
obj4.index.name = 'state'
```
