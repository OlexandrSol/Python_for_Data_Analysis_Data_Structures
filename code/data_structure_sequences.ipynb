### Вбудовані функції послідовностей

#### enumerate

```python
fruits = ['apple', 'banana', 'chery'] 
```

```python
fruits
```

```python
for index, value in enumerate(fruits):
    print(index, value)
```

```python
for index, value in enumerate(fruits, start=1):
    print(index, value)
    
```

#### sorted

```python
sorted([7, 1, 2, 6, 0, 3, 2])
```

```python
sorted('horse race')
```

#### zip

```python
seq1 = ['foo', 'bar', 'baz']
seq2 = ['one', 'two', 'three']
```

```python
zipped = zip(seq1, seq2)
```

```python
zipped
```

```python
list(zipped)
```

```python
seq3 = [False, True]
```

```python
list(zip(seq1, seq2, seq3))
```

```python
for index, (a, b) in enumerate(zip(seq1, seq2)):
    print(f'{index}: {a}, {b}')
```

```python
list(reversed(range(10)))
```
