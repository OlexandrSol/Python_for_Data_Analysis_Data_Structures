```python
a = {1, 2, 3, 4, 5}
b = {3, 4, 5, 6, 7, 8}
```

```python
# Об'єднання множин
a.union(b)
```

```python
a | b
```

```python
# Перетин множин
a.intersection(b)
```

```python
a & b
```

```python
# Різниця
a - b
```

```python
# Додати елемент
a.add(10)
```
