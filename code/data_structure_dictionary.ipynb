#### Значення за змовчуванням

```python
words = ['apple', 'bat', 'bar', 'atom', 'book', 'cat', 'crazy']
```

```python
by_letter = {}
```

```python
for word in words:
    letter = word[0]
    if letter not in by_letter:
        by_letter[letter] = [word]
    else:
        by_letter[letter].append(word)
```

```python
by_letter
```

```python
# За допомогою методу setdefault
by_letter = {}
```

```python
for word in words:
    letter = word[0]
    by_letter.setdefault(letter, []).append(word)
```

```python
by_letter
```

```python
# Через класс defaultdict
from collections import defaultdict
by_letter = defaultdict(list)
```

```python
for word in words:
    by_letter[word[0]].append(word)
```

```python
by_letter
```
