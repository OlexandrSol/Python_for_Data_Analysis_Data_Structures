### Об'єкт DataFrame

```python
import pandas as pd
import numpy as np
```

```python
# Створення DataFrame на основі словника списків
data = {
    'state': ['Ohio', 'Ohio', 'Ohio', 'Nevada', 'Nevada', 'Nevada'], 
    'year': [2000, 2001, 2002, 2001, 2002, 2003],
    'pop': [1.5, 1.7, 3.6, 2.4, 2.9, 3.2]
}
```

```python
data
```

```python
frame = pd.DataFrame(data)
```

```python
frame
```

```python
frame.head()
```

```python
frame.tail()
```

```python
# Можна задати послідовність стовпців
pd.DataFrame(data, columns=['year', 'state', 'pop'])
```

```python
# Якщо запросити відсутній стовпець, то значення будуть NaN
frame2 = pd.DataFrame(data, columns=['year', 'state', 'pop', 'debt'])
```

```python
frame2
```

```python
frame2.columns
```

```python
# Стовпець DataFrame можна вилучити як об'єкт Series
frame2['state']
```

```python
# Доступ до елементів loc та iloc
frame2.loc[2]
```

```python
frame2.iloc[2]
```

```python
# Можна присвоїти скаляр цілому стовпцю
frame2['debt'] = 16.5
```

```python
frame2
```

```python
# Або діапазон
frame2['debt'] = np.arange(6.)
```

```python
frame2
```

```python
# Можна присвоїти Series, але індекси повинні відповідати DataFrame
```

```python
val = pd.Series([-1.2, -1.5, -1.7], index=['two', 'four', 'five'])
```

```python
val
```

```python
frame2['debt'] = val
```

```python
frame2
```

```python
val2 = pd.Series([-1.2, -1.5, -1.7, 0.0, 0.0, 0.0], index=['two', 'four', 'five', 'one', 'three', 'zero'])
```

```python
val2
```

```python
frame2['debt'] = val2
```

```python
frame2
```

```python
val3 = pd.Series([-1.2, -1.5, -1.7, 0.0, 0.0, 0.0], index=[0, 1, 2, 3, 4, 5])
```

```python
frame2['debt'] = val3
```

```python
frame2
```

```python
# Видалення стовпців
frame2['eastern'] = frame2['state'] == 'Ohio'
```

```python
# DataFrame із словника словників
population = {'Ohio': {2000: 1.5, 2001: 1.7, 2002: 3.6}, 
              'Nevada': {2001: 2.4, 2002: 2.9}
             }
```

```python
population
```

```python
frame3 = pd.DataFrame(population)
```

```python
frame3
```

```python
# Об'єкти DataFrame можна транспонувати
frame3.T
```

```python
# Ключі внутрішніх словників об'єднуються для утворення індексу, але якщо індекс заданий
pd.DataFrame(population, index=[2001, 2002, 2003])
```
