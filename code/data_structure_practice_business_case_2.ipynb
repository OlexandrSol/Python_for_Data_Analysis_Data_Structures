```python
from collections import defaultdict
```

```python
# Список словників: кожен словник — це один товар
raw_products_data = [
    {"item": "Ноутбук Acer Aspire 5", "category": " Laptops  ", "brand": "acer"},
    {"item": "Монітор Acer Nitro", "category": "Monitors", "brand": "Acer"},
    {"item": "Ноутбук Acer Swift 3", "category": "laptops", "brand": "ACER "},
    {"item": "Клавіатура Predator", "category": " Accessories", "brand": " predator"},
    {"item": "Ноутбук Predator Helios", "category": "laptops", "brand": "Predator"},
    {"item": "Монітор ConceptD", "category": "Monitors", "brand": "acer"}
]
```

```python
raw_products_data[0]
```

```python
# Задача 1: Стандартизація (Data Cleaning)
```

Створи новий список clean_products, але очисти дані всередині.

Бренд має бути завжди у верхньому регістрі (.upper()) і без зайвих пробілів на початку/в кінці (.strip()).

Категорія має бути завжди з великої літери (.capitalize()) і теж без пробілів.

```python
clean_products = [
    { 'item': x['item'], 
      'category': x['category'].capitalize().strip(),
      'brand': x['brand'].upper().strip()
    }
    for x in raw_products_data]
```

```python
clean_products
```

```python
clean_products = []

for x in raw_products_data:
    # Робимо незалежну копію словника, щоб не псувати raw_products_data
    new_item = x.copy() 
    
    # Працюємо вже з копією
    new_item['brand'] = new_item['brand'].upper().strip()
    new_item['category'] = new_item['category'].strip().capitalize() # Важливо: спочатку strip(), потім capitalize(), щоб прибрати пробіли перед великою літерою
    
    clean_products.append(new_item)
    
print(clean_products)
```

```python
# Задача 2: Групування брендів за категоріями
from collections import defaultdict
```

```python
brands_by_category = defaultdict(set)
```

```python
brands_by_category
```

```python
for x in clean_products:
    brands_by_category[x['category']].add(x['brand'])
```

```python
# Давай виведемо результат у красивому вигляді
for category, brands in brands_by_category.items():
    print(f"Категорія {category} має бренди: {brands}")
```
